```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
normal = np.random.normal(loc=50, scale=10, size=1000)
uniforme = np.random.uniform(low=20, high=80, size=1000)
```

## Análisis de de las estadísticas descriptivas

#### Con Numpy podemos sacar las bases para cualqueir estadística descrptiva, en este caso, de ambas distribuciones generadas.Primero, se halla la media y mediana, las cuales permiten entender dónde se concentra la mayoría de los datos y si hay simetría con la mediana. Despues, calculé la varianza y el desvío estandar, que me indican cuánta dispersión hay en los valores. Esto es clave para ver si los datos están muy separados o agrupados, además de corroborar datos atipicos y por último, con el valor mínimo y máximo se logra tener una idea clara del rango total de cada distribución.


```python
print("Distribución Normal:")
print("Media:", np.mean(normal))
print("Mediana:", np.median(normal))
print("Varianza:", np.var(normal))
print("Desvío estándar:", np.std(normal))
print("Mínimo:", np.min(normal))
print("Máximo:", np.max(normal))

print("\nDistribución Uniforme:")
print("Media:", np.mean(uniforme))
print("Mediana:", np.median(uniforme))
print("Varianza:", np.var(uniforme))
print("Desvío estándar:", np.std(uniforme))
print("Mínimo:", np.min(uniforme))
print("Máximo:", np.max(uniforme))
```

## Visualización de las distribuciones

#### Para visualizar mejor el comportamiento de los datos generados, aproveché Matplotlib y creé dos graficos. En el primer gráfico, mostré el histograma de la distribución normal y el en el segundo gráfico, incluí el histograma de la distribución uniforme, donde los se ve que valores están distribuidos de manera pareja dentro del rango definido.
#### Para que un grafico este correcto se deben tener en cuenta ciertos aspectos los cuales también tuve en cuenta como el agregar títulos, etiquetas en los ejes o leyendas.


```python
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.hist(normal, bins=30, color='skyblue', edgecolor='black', alpha=0.7, label="Normal")
plt.title("Distribución Normal")
plt.xlabel("Valores")
plt.ylabel("Frecuencia")
plt.grid(True)
plt.legend()

plt.subplot(1, 2, 2)
plt.hist(uniforme, bins=30, color='red', edgecolor='black', alpha=0.7, label="Uniforme")
plt.title("Distribución Uniforme")
plt.xlabel("Valores")
plt.ylabel("Frecuencia")
plt.grid(True)
plt.legend()

plt.suptitle("Comparación de Distribuciones Aleatorias")
plt.tight_layout()
plt.show()
```
